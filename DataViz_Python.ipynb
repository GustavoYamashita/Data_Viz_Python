#Projeto Visualização de Dados em Python
### Dados de imigração para p Canadá

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/canadian_immegration_data.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.set_index('Country', inplace = True)
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
brasil = df.loc['Brazil', anos]
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks([str(i) for i in range(1980,2011,5)])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.show()
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df_vendas = pd.DataFrame(vendas_2022, index=lojas)
df_vendas
```

```python
meses = df_vendas.columns
meses
```

```python
fig, axs = plt.subplots(2,2, figsize=(14,6))
fig.subplots_adjust(hspace = 0.4, wspace=0.3)
fig.suptitle('Vendas das lojas A,B,C e D entre jan e dez de 2022')

axs[0,0].plot(df_vendas.loc['A'])
axs[0,0].set_title('A')

axs[0,1].plot(df_vendas.loc['B'])
axs[0,1].set_title('B')

axs[1,0].plot(df_vendas.loc['C'])
axs[1,0].set_title('C')

axs[1,1].plot(df_vendas.loc['D'])
axs[1,1].set_title('D')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Meses')
  ax.set_ylabel('Vendas')

ymin = 0
ymax = 400
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
# Criar a figura e os subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar os espaçamentos entre os subplots
plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Adicionando um título geral para os subplots
fig.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D', fontsize=20)

# Lista de cores
cores = ['darkviolet', 'green', 'darkblue', 'coral']

#Loop para plotar e customizar os subplots
for i, ax in enumerate(axs.flat):
    ax.plot(df_vendas.loc[df_vendas.index[i]], color=cores[i], lw=3)
    ax.set_title(f'Vendas na loja {df.index[i]}', loc='left', fontsize=16)
    ax.set_xlabel('Mês', fontsize=14)
    ax.set_ylabel('Número de vendas', fontsize=14)
    ax.tick_params(labelsize=12)
    ax.grid(color='lightgrey')
```

```python
america_sul = df.query('Region == "South America"')
america_sul
```

```python
cores = ['royalblue','orange','forestgreen','orchid','purple','brown','slateblue','gray','olive','navy','teal','tomato']
fig, ax = plt.subplots(figsize=(15,6))
ax.barh(america_sul.index, america_sul['Total'], color = cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013',loc='left',fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize = 14)
ax.set_ylabel('', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
america_sul_sorted = america_sul.sort_values('Total',ascending=True)

cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brazil':
    cores.append('green')
  else: cores.append('silver')

fig, ax = plt.subplots(figsize=(13,6))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013',loc='left',fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize = 14)
ax.set_ylabel('', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v+20, i, '{:,.0f}'.format(v), color = 'black', fontsize=10, ha='left',va='center')


ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigracao_america_sul.png',transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```
