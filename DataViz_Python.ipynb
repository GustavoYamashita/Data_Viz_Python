#Projeto Visualização de Dados em Python
### Dados de imigração para p Canadá

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/canadian_immegration_data.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.set_index('Country', inplace = True)
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
brasil = df.loc['Brazil', anos]
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks([str(i) for i in range(1980,2011,5)])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'],dados_brasil['imigrantes'], lw=3) #marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize = 18, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.set_ylabel('Número de imigrantes', fontsize=14)
plt.grid(linestyle = "--")
plt.show()
```

```python
with plt.style.context('fivethirtyeight'):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3)
  ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=20, loc='left')
  ax.set_ylabel('Número de imigrantes', fontsize=14)
  ax.set_xlabel('Ano', fontsize=14)
  ax.yaxis.set_tick_params(labelsize=12)
  ax.xaxis.set_tick_params(labelsize=12)
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))
axs[0].plot(dados_brasil['ano'],dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_ylabel('Número de imigrantes')
axs[1].set_xlabel('Brasil')
axs[1].yaxis.set_major_locator(plt.MultipleLocator(250))
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace = 0.5, wspace=0.3)
fig.suptitle('Imigração dos 4 maiores países da América do Sul para o Canadá 1980 a 2013')

axs[0,0].plot(df.loc['Brazil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colombia', anos])
axs[0,1].set_title('Colombia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

ymin = 0
ymax = 7000
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df = pd.DataFrame(vendas_2022, index=lojas)
df
```

```python
meses = df.columns
meses
```

```python
fig, axs = plt.subplots(2,2, figsize=(14,6))
fig.subplots_adjust(hspace = 0.4, wspace=0.3)
fig.suptitle('Vendas das lojas A,B,C e D entre jan e dez de 2022')

axs[0,0].plot(df.loc['A'])
axs[0,0].set_title('A')

axs[0,1].plot(df.loc['B'])
axs[0,1].set_title('B')

axs[1,0].plot(df.loc['C'])
axs[1,0].set_title('C')

axs[1,1].plot(df.loc['D'])
axs[1,1].set_title('D')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Meses')
  ax.set_ylabel('Vendas')

ymin = 0
ymax = 400
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```
