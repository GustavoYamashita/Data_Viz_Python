```python
import pandas as pd
```

```python
df = pd.read_csv('/content/canadian_immegration_data.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.set_index('Country', inplace = True)
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
brasil = df.loc['Brazil', anos]
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks([str(i) for i in range(1980,2011,5)])
plt.title('ImigraÃ§Ã£o do Brasil para o CanadÃ¡')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de imigrantes')
plt.show()
```

```python
bra_arg = df.loc[['Brazil','Argentina'], anos]
bra_arg = bra_arg.T
#bra_arg_dict = {'ano': bra_arg.index.tolist(), 'imigrantes': bra_arg.values.tolist()}
#dados_bra_arg = pd.DataFrame(bra_arg_dict)
#dados_bra_arg
bra_arg
```

```python
plt.figure(figsize=(8,4))
plt.plot(bra_arg['Argentina'], label = 'Argentina')
plt.plot(bra_arg['Brazil'], label = 'Brasil')
plt.xticks([str(i) for i in range(1980,2011,5)])
plt.title('ImigraÃ§Ã£o do Brasil e Argentina para o CanadÃ¡')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de imigrantes')
plt.legend()
plt.show()
```
